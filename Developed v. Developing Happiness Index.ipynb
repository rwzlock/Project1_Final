```python
import pandas as pd
import csv
import matplotlib.pyplot as plt

from pylab import rcParams
rcParams['figure.figsize'] = 12, 7
```

```python
groupby_data = pd.read_csv("Groupby_EconomyLevel_Dataset.csv")
groupby_data
```

```python
#y
developed_social_data = groupby_data.loc[groupby_data['Economy_Level']=='Developed',:].groupby(['Economy_Level','country']).mean()['Social support']
developing_social_data = groupby_data.loc[groupby_data['Economy_Level']=='Developing',:].groupby(['Economy_Level','country']).mean()['Social support']
#x
developed_lifeladder_data = groupby_data.loc[groupby_data['Economy_Level']=='Developed',:].groupby(['Economy_Level','country']).mean()['Life Ladder']
developing_lifeladder_data = groupby_data.loc[groupby_data['Economy_Level']=='Developing',:].groupby(['Economy_Level','country']).mean()['Life Ladder']
```

```python
#Social support v. Happiness score
plt.scatter(developed_lifeladder_data,developed_social_data, color = "springgreen", label = "Developed",s=60, edgecolors='none')
plt.scatter(developing_lifeladder_data,developing_social_data, color = "b", label = "Developing", s=60,edgecolors='none')

plt.xlim(0,10)
plt.ylim(-0.05,1)
plt.title("Social Support & Happiness",fontsize=20)
plt.xlabel("Happiness Score",fontsize=20)
plt.ylabel("Social Support Score",fontsize=20)
plt.legend()
plt.grid()

plt.savefig("Happy_SosialSupport.png")
plt.show()
```

```python
#Log GDP v. Happiness Score
developed_gdp_data = groupby_data.loc[groupby_data['Economy_Level']=='Developed',:].groupby(['Economy_Level','country']).mean()['Log GDP per capita']
developing_gdp_data = groupby_data.loc[groupby_data['Economy_Level']=='Developing',:].groupby(['Economy_Level','country']).mean()['Log GDP per capita']

plt.scatter(developed_lifeladder_data,developed_gdp_data, color = "springgreen", label = "Developed",s=60,edgecolors='none')
plt.scatter(developing_lifeladder_data,developing_gdp_data, color = "b", label = "Developing",s=60,edgecolors='none')

plt.xlim(0,10)
plt.title("Log GDP per Capita & Happiness",fontsize=20)
plt.xlabel("Happiness Score",fontsize=20)
plt.ylabel("Log GDP per Capita",fontsize=20)

plt.legend()
plt.grid()

plt.savefig("Happy_gdp.png")
plt.show()
```

```python
#Healthy Life Expectancy at Birth v. Happiness Score
developed_health = groupby_data.loc[groupby_data['Economy_Level']=='Developed',:].groupby(['Economy_Level','country']).mean()['Healthy life expectancy at birth']
developing_health = groupby_data.loc[groupby_data['Economy_Level']=='Developing',:].groupby(['Economy_Level','country']).mean()['Healthy life expectancy at birth']

plt.scatter(developed_lifeladder_data,developed_health, color = "springgreen", label = "Developed",s=60,edgecolors='none')
plt.scatter(developing_lifeladder_data,developing_health, color = "b", label = "Developing",s=60,edgecolors='none')
plt.xlim(0,10)
plt.ylim(30,80)
plt.title("Healthy Life Expectancy at Birth & Happiness",fontsize=20)
plt.xlabel("Happiness Score",fontsize=20)
plt.ylabel("Healthy Life Expectancy at Birth Score",fontsize=20)

plt.legend()
plt.grid()

plt.savefig("Happy_health.png")
plt.show()
```
